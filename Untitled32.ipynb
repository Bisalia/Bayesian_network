1. Промоделировать нелинейную модель y = f(x, a, b) + δ с несмещенной нормально распределенной
ошибкой, дисперсия которой равна ε, считая x стандартно нормально распределенной случайной
величиной. В качестве примера рассматривается последний вариант, поэтому вариантов всего
шесть, (g) исключено.

```python
import numpy as np
import matplotlib.pyplot as plt

# Установить параметры шаблона
a = 0.4
b = 1
epsilon = 0.3

# Генерировать случайные данные для x
x = np.random.normal(0, 1, size=100)
# Вычислить у Для каждого значения х с использованием шаблона
y = b + np.exp(a*x) + np.random.normal(0, np.sqrt(epsilon), size=100)

# Оценка параметров а и в с использованием обычного метода наименьших квадратов
A = np.vstack([np.exp(a*x), np.ones(len(x))]).T
a_hat, b_hat = np.linalg.lstsq(A, y, rcond=None)[0]

# Предсказать значение у для нового значения х
x_new = 2
y_new = b_hat + np.exp(a_hat*x_new)
# Рассчитать коэффициент детерминации (R2)
y_pred = b_hat + np.exp(a_hat*x)
SSR = np.sum((y_pred - np.mean(y))**2)
SSTO = np.sum((y - np.mean(y))**2)
R2 = SSR / SSTO

# Показать результаты
plt.scatter(x, y)
plt.plot(x, b_hat + np.exp(a_hat*x), 'r', label='Fit')
plt.plot(x_new, y_new, 'go', label=f'Prediction ({x_new:.2f}, {y_new:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Model Fit (R²={R2:.2f})')
plt.legend()
plt.show()
```

```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Définir les paramètres du modèle
a = 0.4
b = 1
epsilon = 0.3

# Générer des données aléatoires pour x
x = np.random.normal(0, 1, size=100)

# Calculer y pour chaque valeur de x en utilisant le modèle
y = b + np.exp(a*x) + np.random.normal(0, np.sqrt(epsilon), size=100)

# Définir la fonction non linéaire à ajuster
def nonlinear_func(x, a, b):
    return b + np.exp(a*x)

# Estimer les paramètres a et b en utilisant la méthode des moindres carrés
popt, pcov = curve_fit(nonlinear_func, x, y)

# Prédire la valeur de y pour une nouvelle valeur de x
x_new = 2
y_new = nonlinear_func(x_new, *popt)

# Calculer le coefficient de détermination (R²)
y_pred = nonlinear_func(x, *popt)
SSR = np.sum((y_pred - np.mean(y))**2)
SSTO = np.sum((y - np.mean(y))**2)
R2 = SSR / SSTO

# Afficher les résultats
plt.scatter(x, y)
plt.plot(x, nonlinear_func(x, *popt), 'r', label='Nonlinear Fit')
plt.plot(x_new, y_new, 'go', label=f'Prediction ({x_new:.2f}, {y_new:.2f})')
plt.xlabel('x')
plt.ylabel('y')
plt.title(f'Nonlinear Model Fit (R²={R2:.2f})')
plt.legend()
plt.show()
```

2. Оценить параметры нелинейной модели по методу наименьших квадратов (численно). Применить
к модельным данным линейную модель и оценить параметры. Построить на двумерной диаграмме
основную и линейную модель. Сравнить невязки для обеих моделей.


```python
from scipy.optimize import leastsq
import numpy as np
```

```python
#Данные
x_data =  np.array([1,2,3,4,5])
y_data =  np.array([2.1, 3.8, 5.3, 6.5, 8.2])
```

```python
#Нелинейная функция
def f_nonlinear(x, a, b):
  return b + np.exp(x*a)
```

```python
# Функция наименьших квадратов для нелинейной модели
def residuals_nonlinear(params,y, x):
  a, b = params
  return y-f_nonlinear(x, a,b)
```

```python
# Начальные параметры
params_init_nonlinear = np.array([1,0.8])
```

```python
# Подгонка нелинейной модели
params_opt_nonlinear, _ = leastsq(residuals_nonlinear, params_init_nonlinear, args=(x_data, y_data))
```

```python
#Отображение параметров нелинейной модели
print('a_nonlin=', params_opt_nonlinear[0])
print('b_nonlin=', params_opt_nonlinear[1])
```

```python
# Расчет значений нелинейной модели
y_model_nonlinear = f_nonlinear(x_data, params_opt_nonlinear[0], params_opt_nonlinear[1])
```

```python
#линейная функция
def f_linear(x, a, b):
  return a*x + b
```

```python
# Функция наименьших квадратов для линейной модели
def residuals_linear(params, x, y):
  a, b = params
  return y - f_linear(x, a, b)
```

```python
# Начальные параметры
params_init_linear = np.array([1, 0.8])
```

```python

# Подгонка линейной модели
params_opt_linear, _ = leastsq(residuals_linear, params_init_linear, args=(x_data, y_data))
```

```python
#Отображение параметров линейной модели
print('a_lin =', params_opt_linear[0])
print('b_lin =', params_opt_linear[1])
```

```python
# Расчет значений нелинейной модели
y_model_linear = f_linear(x_data, params_opt_linear[0], params_opt_linear[1])
```

```python
#Построение данных и моделей
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(x_data, y_data, 'o', label='Données')
ax.plot(x_data, y_model_nonlinear, label='Modèle non linéaire')
ax.plot(x_data, y_model_linear, label='Modèle linéaire')
ax.legend()
plt.show()
```

```python
#Расчет остаточных значений для обеих моделей
residuals_nonlinear = y_data - y_model_nonlinear
residuals_linear = y_data - y_model_linear
```

```python
print('Valeurs résiduelles du modèle non linéaire :', residuals_nonlinear)
print('Valeurs résiduelles du modèle linéaire :', residuals_linear)
```
