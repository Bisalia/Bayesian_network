```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import pearsonr, spearmanr

```

```python
df  = pd.read_excel("life0.xlsx")
df
```

```python
A =df[["чиновн","самоуб","импорт"]]
```

```python
sns.pairplot(A)
```

```python
pearson = A.corr(method="pearson")
pearson
```

```python
print("la premier relation :", pearsonr(A["чиновн"],A["самоуб"]))
print("la premier relation :", pearsonr(A["чиновн"],A["импорт"]))
print("la premier relation :", pearsonr(A["самоуб"],A["импорт"]))
```

По коэффициенту Пирсона корреляция значима между переменными: (доходы бедных,водка) и (доходы бедных, чиновники) на уровне значимости 0.1

```python
spearman = A.corr(method="spearman")
spearman 
```

```python
print("la premier relation :", spearmanr(A["чиновн"],A["самоуб"]))
print("la premier relation :", spearmanr(A["чиновн"],A["импорт"]))
print("la premier relation :", spearmanr(A["самоуб"],A["импорт"]))
```

По коэффициенту Спирмана значимых корреляций нет

```python
coef = A.corr(method='kendall')
coef 
```

```python
sns.heatmap(coef.corr());
```
